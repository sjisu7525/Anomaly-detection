```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, roc_curve
import pandas as pd

def score(true, pred) :
  acc = accuracy_score(true,pred)
  f1 = f1_score(true,pred)
  recall = recall_score(true,pred)
  precision = precision_score(true,pred)

  df = pd.DataFrame([acc,f1,recall,precision], index = ['Acc','F1-score','Recall','Precision'], columns = ['value'])

  return df
```
