```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

def remove_trend(data) :
  length = len(data)

  x = list(range(0,length))
  line = pd.DataFrame(list(zip(x, data)))

  line_fitter = LinearRegression()
  line_fitter.fit(line.iloc[:,0].values.reshape(-1,1), line.iloc[:,1])
  y_predicted = line_fitter.predict(line.iloc[:,0].values.reshape(-1,1))

  data_ = data - y_predicted

  return data_
```
