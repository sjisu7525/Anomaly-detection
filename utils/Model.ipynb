```python
from keras.layers import Input, Dense, BatchNormalization
from keras.models import Model

def autoencoder(window, dims) :

  input_data = Input(shape=(window,))

  encoded = Dense(dims[0], activation='linear')(input_data)

  for i in range(1, len(dims)) :
    encoded = BatchNormalization()(encoded)
    encoded = Dense(dims[i], activation='relu')(encoded)

  decoded = Dense(dims[-2], activation='relu')(encoded)

  for i in reversed(range(0, len(dims)-2)) :
    decoded = BatchNormalization()(decoded)
    decoded = Dense(dims[i], activation='relu')(decoded)

  decoded = BatchNormalization()(decoded)
  decoded = Dense(window, activation = 'linear')(decoded)

  autoencoder = Model(input_data, decoded)
  encoder = Model(input_data, encoded)

  autoencoder.compile(optimizer='adam', loss='mse', metrics=['mse']) # model.compile(loss= 'mse', optimizer= Adam(lr= 0.01))

  return autoencoder, encoder
```
