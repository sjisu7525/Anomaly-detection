```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import utils
from utils.find_window import window
from utils.remove_trend import remove_trend
from utils.decompose import decompose
from utils.Model import autoencoder
from utils.train_ae import training
from utils.score import score
```

```python
def detect_anomaly(data, lags, ae_dims, list_ae, stride, fig = True) :

  # data.iloc[:,0] : 시간에 따른 관측값
  # data.iloc[:,1] : 이상값 여부 true label
  # lags : pacf_그림의 차수
  # list_ae : [epoch1, batch1, epoch2, batch2]
  # stride : stride of frequency domain
  # dims : dims of autoencoder

  ## window size
  window_size = window(data.iloc[:,0], lags = lags)

  ## remove trend
  data_ = remove_trend(data.iloc[:,0])

  ## decompose
  cycle, trend = decompose(data_, lamda = 100)

  ## training
  ae = autoencoder
  epoch1, batch1, epoch2, batch2 = list_ae
  data, RE_final_test_cycle, RE_final_test_trend, anomaly_idx = training(data, cycle, trend, window_size, ae, ae_dims, epoch1, batch1, epoch2, batch2, stride)

  ## score
  score_df = score(data.iloc[:, 1], data.iloc[:, 2])
  print(score_df)

  ## plot
  if fig :
    fig = plt.figure(figsize=(15,5))
    plt.plot(data.iloc[:,0])
    plt.plot(data.iloc[anomaly_idx, 0],'ro', markersize = 4)

  return data, RE_final_test_cycle, RE_final_test_trend, score_df
```
